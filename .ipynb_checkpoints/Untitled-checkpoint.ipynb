```python
my_min, my_max = apple.get_min_max(gaussian_values)
x = apple.df.index
print(x[my_max])
print(x[my_min])
```

```python
import numpy as np
i = 0
diff_arr = []
while (i < min([len(x[my_max]), len(x[my_min])])):
    diff_arr.append((x[my_max][i]-x[my_min][i]).days)
    print(x[my_max][i]-x[my_min][i])
    i+=1
    
print(np.mean(diff_arr))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
plt.plot(diff_arr, '-ok', color='black');
```
